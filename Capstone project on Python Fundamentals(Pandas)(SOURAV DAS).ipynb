##Task 2
The cost column in dataframe “Project” has some missing values. Your task is to
compute these missing values. Replace the missing values by running average. You
should use “For” loop for this task.

```python
# reading csv file
pd.read_csv("Projects_Data_Table.csv", header = [0,1])
```

```python
# Using For Loop for the range of row numbers of the DataFrame
for i in range(0,Projects_Data_Table.shape[0]):
    
# replacing the missing or NaN values by the running average values using the rolling().mean() function
    Projects_Data_Table[('Project DataFrame','Cost')]=round(Projects_Data_Table[('Project DataFrame','Cost')].fillna(Projects_Data_Table[('Project DataFrame','Cost')].rolling(3,min_periods=1).mean()),0)

# saving the updated dataframe as a new .csv file
Projects_Data_Table.to_csv('Projects_Data_New_Table.csv', index=False)

# checking dataframe
Projects_Data_Table
```

##Task 3 Split the name column in Employee dataframe to two new columns “First Name”, “Last
Name” and remove the older “name” column.

```python
# Reading the Employee Dataframe from the saved CSV file by keeping the multiple headers and the empty values
Employee_Data_Table = pd.read_csv('Employee_Data_Table.csv', header = [0,1], na_filter = False)

# Inserting a new column First Name at column index = 1
Employee_Data_Table.insert(1,('Employee DataFrame','First Name'),'')

# Inserting the First Name values by splitting the name column
Employee_Data_Table['Employee DataFrame','First Name'] = Employee_Data_Table['Employee DataFrame','Name'].str.split(' ').str[0]

# Inserting a new column Last Name at column index = 2
Employee_Data_Table.insert(2,('Employee DataFrame','Last Name'),'')

# Inserting the Last Name values by splitting the name column
Employee_Data_Table['Employee DataFrame','Last Name'] = Employee_Data_Table['Employee DataFrame','Name'].str.split(' ').str[1]

# Dropping the name column from the Employee Dataframe
Employee_Data_Table = Employee_Data_Table.drop(['Name'], axis = 1, level = 1)

# Saving the updated Employee Dataframe as a new CSV file
Employee_Data_Table.to_csv('Employee_Data_New_Table.csv', index = False)

# checking dataframe
Employee_Data_Table
```

##Task 4 Join all three dataframes in one single dataframe. Name it “Final”

```python
# Dropping a level down
Employee_Data_Table.columns = Employee_Data_Table.columns.droplevel(0)

# checking dataframe
Employee_Data_Table
```

```python
# Dropping a level down
Seniority_Level_Data_Table.columns = Seniority_Level_Data_Table.columns.droplevel(0)

# checking dataframe
Seniority_Level_Data_Table
```

```python
# Dropping a level down
Projects_Data_Table.columns = Projects_Data_Table.columns.droplevel(0)

# checking dataframe
Projects_Data_Table
```

```python
# Joining all three dataframes in one single dataframe Final
Final = pd.merge(pd.merge(Employee_Data_Table, Seniority_Level_Data_Table, on ='ID'), Projects_Data_Table, on ='ID')

# checking dataframe
Final
```

##Task 5 Add a new bonus column in Final dataframe. Give 5% bonus concerning project cost
  only for employees who have finished the projects.


```python
# Using For Loop for the range of row numbers of the DataFrame
for i in range(Final.shape[0]):
    
# Checking the conditions under IF statement
    if Final.loc[i,'Status'] == 'Finished':
        
# Doing the corresponding calculations           
        Final.loc[i,'Bonus'] = Final.loc[i,'Cost']*.05        
```

```python
# Using fillna() function for the whole dataframe
Final = Final.fillna(0)

# checking dataframe
Final
```

##Task 6 Decrease the designation level by 1, whose projects have status “fail”. Delete the
employee record whose designation level is below 4. 

```python
# Identifying all the unique IDs
for i in Final.ID.unique():
    
# For each unique ID, checking if there are any projects with a Failed Status   
    if (Final.loc[Final.ID == i,'Status'] == 'Failed').any() == True:
        
# Decreasing the Designation Level  (numerically increasing) of the employee with projects with a Failed Status            
        Final.loc[Final.ID == i,'Designation Level'] += 1
```

```python
# checking dataframe
Final
```

```python
# Using For Loop for the range of row numbers of the DataFrame
for i in range(Final.shape[0]):
    
# Checking the conditions under IF statement    
    if Final.loc[i,'Designation Level'] > 4:
        
# Dropping the row of the employee record whose designation level is below 4.       
        Final.drop([i], inplace = True)
```

```python
# checking dataframe
Final
```

##Task 7 Add a new column for gender, than add “Mr.” and “Mrs” to the first name column.


```python
# Declaring a list that is to be converted into a column
Gender = ['M', 'M', 'M', 'F', 'F', 'M', 'M', 'M', 'F', 'F', 'F', 'F', 'F', 'F']
 
# Inserting a new column Gender at column index = 5
Final.insert(5,('Gender'),'')

# Using Gender as the column name and equating it to the list
Final['Gender'] = Gender

# checking dataframe
Final
```

```python
# Using For Loop for the range of row numbers of the DataFrame
for i in range(Final.shape[0]):
    
# Checking the conditions under IF-ELSE statement and doing the corresponding task    
    if Final.loc[i,'Gender'] == 'M':
        Final.loc[i,'First Name'] = 'Mr.' + Final.loc[i,'First Name']
        
    else: 
        Final.loc[i,'First Name'] = 'Mrs ' + Final.loc[i,'First Name']   
```

```python
# checking dataframe
Final
```

##Task 8 Increase designation level by 1 for the employees whose age is more than 29 years using
IF condition


```python
# Using For Loop for the range of row numbers of the DataFrame
for i in range(Final.shape[0]):

# Checking the conditions under IF statement     
    if Final.loc[i,'Age'] > 29:
        
# Increasing the Designation Level  (numerically decreasing) of the employee with age more than 29 years              
        Final.loc[i,'Designation Level'] = Final.loc[i,'Designation Level']-1
```

```python
# checking dataframe
Final
```

##Task 9 Add the cost of all projects for each Employee and save it in new dataframe
“TotalProjCost” with three columns ID, First Name, and Total cost


```python
# Creating a New DataFrame of specific columns
NDF = pd.DataFrame()
NDF['ID'] = Final['ID']
NDF['First Name'] = Final['First Name']
NDF['Total Cost'] = Final['Cost']
```

```python
# Using GroupBy multiple column
TotalProjCost = NDF.groupby(['ID', 'First Name']).sum()

# checking dataframe
TotalProjCost
```
